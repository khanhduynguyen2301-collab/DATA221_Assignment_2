```python
import pandas as pd

student_file_read = pd.read_csv('student.csv')

grade_band_list = []
for grade in student_file_read['grade']:
    if grade >= 15:
        grade_check = 'High'
    elif 14 >= grade >= 10:
        grade_check = 'Medium'
    else:
        grade_check = 'Low'

    grade_band_list.append(grade_check)

student_file_read['grade_band'] = grade_band_list

summary_table = (student_file_read.groupby("grade_band").agg(number_of_students=("grade", "count"),
                                             average_absences=("absences", "mean"),
                                            internet_access_percentage=("internet", "mean")).reset_index())

summary_table["internet_access_percentage"] = summary_table["internet_access_percentage"] * 100
summary_table.to_csv("student_bands.csv", index = False)
print('Saved successfully!')
print()
print(summary_table)
```
