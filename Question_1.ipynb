```python
punctuation = ".,!?;:\"'()[]{}<>â€”-"
cleaned_words = []

with open('sample-file.txt', 'r', encoding="utf-8") as file:
    for token in file.read().split():
        word = token.lower()

        while len(word) > 0 and word[0] in punctuation:
            word = word[1:]
        while len(word) > 0 and word[-1] in punctuation:
            word = word[:-1]

        alpha_count = 0
        for char in word:
            if char.isalpha():
                alpha_count += 1

        if alpha_count >= 2:
            cleaned_words.append(word)

# Build frequency dictionary
word_frequency_dictionary = {}

for word in cleaned_words:
    if word in word_frequency_dictionary:
        word_frequency_dictionary[word] += 1
    else:
        word_frequency_dictionary[word] = 1

# ---- MANUALLY find top 10 words ----
already_printed_list = []

for i in range(10):
    max_word = None
    max_count = 0

    for word, count in word_frequency_dictionary.items():
        if word not in already_printed_list and count > max_count:
            max_word = word
            max_count = count

    if max_word is None:
        break   # no more words left

    print(f"{max_word} -> {max_count}")
    already_printed_list.append(max_word)
```
