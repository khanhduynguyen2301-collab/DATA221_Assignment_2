```python
def find_lines_containing(filename, keyword):
    """
    Returns a list of (line_number, line_text) for lines that contain ←↩
        keyword
    (case-insensitive). Line numbers start at 1.
    """
    matching_lines = []

    with open(filename, "r") as file:
        for line_number, line in enumerate(file, start = 1):
            if keyword.lower() in line.lower():
                matching_lines.append((line_number, line.strip()))
    return matching_lines

result = find_lines_containing('sample-file.txt', keyword = 'lorem')

print(f"The number of matching lines: {len(result)}")
if len(result) == 0:
    print("No matching lines found.")

elif 1 <= len(result) <= 3:
    print(f'The first {len(result)} matching lines are:')
    for line_number, line in result:
        print(line_number, line)

else:
    print(f'The first 3 matching lines are:')
    for line_number, line in result[:3]:
        print(line_number, line)
```
