```python
import pandas as pd


crime_file_read = pd.read_csv('crime.csv')

risk_list = ["High-Crime" if item >= 0.5 else 'LowCrime' for item in crime_file_read['ViolentCrimesPerPop']]

crime_file_read['risk'] = risk_list
avg_unemployment = (crime_file_read.groupby("risk")["PctUnemployed"].mean())

print("Average Unemployment Rate by Crime Risk Category:\n")
for group, value in avg_unemployment.items():
    print(f"{group}: {value:.2f}%")
```
