```python
import pandas as pd


crime_file_read = pd.read_csv('crime.csv')

risk_list = ["HighCrime" if item >= 0.5 else 'LowCrime' for item in crime_file_read['ViolentCrimesPerPop']]

crime_file_read['risk'] = risk_list
average_unemployment = (crime_file_read.groupby("risk")["PctUnemployed"].mean())

print("Average Unemployment Rate by Crime Risk Category:\n")
for group, value in average_unemployment.items():
    print(f"{group}: {value}")
```
