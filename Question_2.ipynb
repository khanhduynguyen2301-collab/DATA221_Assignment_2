```python
punctuation = ".,!?;:\"'()[]{}<>â€”-"
cleaned_words = []

with open('sample-file.txt', 'r', encoding = "utf-8") as file:
    for token in file.read().split():
        word = token.lower()

        while len(word) > 0 and word[0] in punctuation:
            word = word[1:]
        while len(word) > 0 and word[-1] in punctuation:
            word = word[:-1]

        alpha_count = 0
        for char in word:
            if char.isalpha():
                alpha_count += 1

        if alpha_count >= 2:
            cleaned_words.append(word)

# Build bigrams
bigrams_list = []

for index in range(len(cleaned_words) - 1):
    pair = cleaned_words[index] + ' ' + cleaned_words[index + 1]
    bigrams_list.append(pair)

# Build frequency dictionary
bigrams_frequency_dictionary = {}

for bigram in bigrams_list:
    if bigram in bigrams_frequency_dictionary:
        bigrams_frequency_dictionary[bigram] += 1
    else:
        bigrams_frequency_dictionary[bigram] = 1

# ---- MANUALLY find top 5 bigrams ----
already_printed_list = []

for i in range(5):
    max_bigram = None
    max_count = 0

    for bigram, count in bigrams_frequency_dictionary.items():
        if bigram not in already_printed_list and count > max_count:
            max_bigram = bigram
            max_count = count

    if max_bigram is None:
        break   # no more bigrams left

    print(f"{max_bigram} -> {max_count}")
    already_printed_list.append(max_bigram)
```
