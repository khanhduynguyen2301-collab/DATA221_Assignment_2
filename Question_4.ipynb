```python
import pandas as pd

student_file_read = pd.read_csv('student.csv')

filtered_df = student_file_read[
    (student_file_read["studytime"] >= 3) &
    (student_file_read["internet"] == 1) &
    (student_file_read["absences"] <= 5)]

filtered_df.to_csv('high_engagement.csv', index = False)

num_students = len(filtered_df)
avg_grade = filtered_df["grade"].mean()

print(f"Number of students saved: {num_students}")
print(f"Average grade of these students: {avg_grade:.2f}")

```
